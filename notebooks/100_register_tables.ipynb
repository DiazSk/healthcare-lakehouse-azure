```python
from utils.paths import GOLD

gold_tables = [
    "dim_provider", 
    "dim_hcpcs", 
    "dim_geography",
    "fact_provider_service",
    "gold_hero_geo_arbitrage", 
    "gold_hero_non_par_premium",
    "gold_hero_jcode_concentration", 
    "gold_hero_credentials_markup",
    "gold_hero_site_neutral"
]

for table in gold_tables:
    table_path = GOLD(table)
    # This maps the Azure Delta files to the Databricks SQL endpoint
    spark.sql(f"DROP TABLE IF EXISTS hive_metastore.default.{table}")
    spark.sql(f"CREATE TABLE hive_metastore.default.{table} USING DELTA LOCATION '{table_path}'")
    print(f"\u2705 Successfully registered: hive_metastore.default.{table}")
```
